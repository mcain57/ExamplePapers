```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
#initialisation
r1a = [1,2,3]
r1b = [4,5,6]
r2a = [2,3,2]
r2b = [5,6,7]
```

```python
def distance(r1a, r1b, r2a, r2b):
    r1a = np.array(r1a)
    r2a = np.array(r2a)
    r1b = np.array(r1b)
    r2b = np.array(r2b)
    cross = np.cross(r1b, r2b)
    d = (np.linalg.norm(np.dot((r2a-r1a),cross)))/(np.linalg.norm(cross))
    return d.item()
```

```python
distance(r1a, r1b, r2a, r2b)
```
